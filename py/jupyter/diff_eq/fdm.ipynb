# Метод конечных элементов, Finite difference method

## это численный метод для решения дифференциальных уравнений, который заменяет производные разностными аналогами. Вот примеры реализации для разных типов уравнений с использованием NumPy и SciPy.

## 1. Простое одномерное уравнение (уравнение Пуассона)

Рассмотрим уравнение:

u''(x) = f(x), x ∈ [a, b]
с граничными условиями u(a) = α, u(b) = β,

## Решение

Разбиваем отрезок [a, b] на n узлов, и нумеруем их от 0 до (n-1). Тогда узлы на границах и шаг сетки: $$u_{0}=α, u_{n-1}=β, h=\frac{(b-a)}{(n-1)}$$
а во внутренней области для узлов от 1 до (n-2) аппроксимируем производную второго порядка центральной конечной разностью:
$$
\frac{u_{i-1}-2u_{i}+u_{i+1}}{h^2}=f(x_{i})
$$
т.е.:
$$
\frac{1}{h^2}*u_{i-1}-\frac{2}{h^2}*u_{i}+\frac{1}{h^2}*u_{i+1}=f(x_{i}), i ∈ [1,n-2]
$$
$$
u_{0}=α, u_{n-1}=β
$$
для решения СЛАУ необходимо привести к виду Au=f, где A - квадратная матрица коэффициентов или в матричной форме,

$$
\frac{1}{h^2} \,
\begin{vmatrix}
1 & 0 & 0 & 0 & 0  & 0 & ...\\
1 & -2 & 1 & 0 & 0 & 0 & ...\\
0 & 1 & -2 & 1 & 0 & 0 & ...\\
0 & 0 & 1 & -2 & 1 & 0 & ...\\
... \\
0 & 0 & ... 1 & -2 & 1 & 0 & 0 \\
0 & 0 & ... 0 & 1 & -2 & 1 & 0 \\
0 & 0 & ... 0 & 0 & 1 & -2 & 1 \\
0 & 0 & ... 0 & 0 & 0 & 0 & 1
\end{vmatrix}
\begin{vmatrix}
u_{0} \\
u_{1} \\
u_{2} \\
...   \\
...   \\
...   \\
u_{n-3} \\
u_{n-2} \\
u_{n-1}
\end{vmatrix} =
\begin{vmatrix}
f_{0} \\
f_{1} \\
f_{2} \\
...   \\
...   \\
...   \\
f_{n-3} \\
f_{n-2} \\
f_{n-1}
\end{vmatrix}
$$
где
$$
u_{0}=f_{0}=\alpha, u_{n-1}=f_{n-1}=\beta, f_{i} = f(x_{i}) \, для \ i \in [1, (n-2)] 
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def solve_poisson(f, a, b, alpha, beta, n):
    """
    Решение уравнения u''(x) = f(x) с граничными условиями u(a)=alpha, u(b)=beta
    методом конечных разностей.
    
    Параметры:
        f - функция правой части
        a, b - границы интервала
        alpha, beta - граничные условия
        n - количество точек сетки (включая границы)
    """
    h = (b - a) / (n - 1)  # шаг сетки
    x = np.linspace(a, b, n)  # создание сетки
    
    # Инициализация матрицы и правой части
    A = np.zeros((n, n))
    F = np.zeros(n)
    
    # Граничные условия
    A[0, 0] = 1
    F[0] = alpha
    A[-1, -1] = 1
    F[-1] = beta
    
    # Внутренние точки
    for i in range(1, n-1):
        A[i, i-1] = 1 / h**2
        A[i, i] = -2 / h**2
        A[i, i+1] = 1 / h**2
        F[i] = f(x[i])

    # Решение системы
    u = np.linalg.solve(A, F)
    
    return x, u
```

```python
# Пример использования
def f(x):
    return -np.pi**2 * np.sin(np.pi * x)

a, b = 0, 1
alpha, beta = 0, 0
n = 100

x, u = solve_poisson(f, a, b, alpha, beta, n)

# Аналитическое решение для сравнения
u_exact = np.sin(np.pi * x)

plt.plot(x, u, label='Численное решение')
plt.plot(x, u_exact, '--', label='Точное решение')
plt.grid(True)
plt.legend()
plt.xlabel('x')
plt.ylabel('u(x)')
plt.title('Решение уравнения Пуассона')
plt.show()
```

## 2. Двумерное уравнение Пуассона

Для уравнения:

∇²u = f(x,y) в области Ω с граничными условиями u|∂Ω = g(x,y)

```python
from scipy.sparse import diags, eye
from scipy.sparse.linalg import spsolve
```

```python
def solve_poisson_2d(f, g, xmin, xmax, ymin, ymax, nx, ny):
    """
    Решение 2D уравнения Пуассона методом конечных разностей.
    """
    hx = (xmax - xmin) / (nx - 1)
    hy = (ymax - ymin) / (ny - 1)
    
    x = np.linspace(xmin, xmax, nx)
    y = np.linspace(ymin, ymax, ny)
    X, Y = np.meshgrid(x, y)
    
    # Инициализация правой части и решения
    F = f(X, Y)
    U = np.zeros((ny, nx))
    
    # Применение граничных условий
    U[0, :] = g(x, ymin)
    U[-1, :] = g(x, ymax)
    U[:, 0] = g(xmin, y)
    U[:, -1] = g(xmax, y)
    
    # Построение матрицы (используем разреженные матрицы для эффективности)
    N = (nx - 2) * (ny - 2)
    main_diag = -2/hx**2 - 2/hy**2 * np.ones(N)
    x_diag = (1/hx**2) * np.ones(N - (ny - 2))
    y_diag = (1/hy**2) * np.ones(N - 1)
    
    diagonals = [main_diag, x_diag, x_diag, y_diag, y_diag]
    A = diags(diagonals, [0, (ny - 2), -(ny - 2), 1, -1], format='csr')
    
    # Правая часть (учитываем граничные условия)
    F_int = F[1:-1, 1:-1].flatten()
    
    # Решение системы
    U_int = spsolve(A, F_int)
    U[1:-1, 1:-1] = U_int.reshape((ny - 2, nx - 2))
    
    return X, Y, U
```

```python
# Пример использования
def f(x, y):
    return -2 * np.pi**2 * np.sin(np.pi * x) * np.sin(np.pi * y)

def g(x, y):
    return np.zeros_like(x)

xmin, xmax = 0, 1
ymin, ymax = 0, 1
nx, ny = 50, 50

X, Y, U = solve_poisson_2d(f, g, xmin, xmax, ymin, ymax, nx, ny)

# Визуализация
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, U, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('u(x,y)')
ax.set_title('Решение 2D уравнения Пуассона')
plt.show()
```

## 3. Уравнение теплопроводности (параболическое уравнение)

Рассмотрим:

∂u/∂t = α ∂²u/∂x², x ∈ [0, L], t ∈ [0, T]
с начальным условием u(x,0) = u0(x)
и граничными условиями u(0,t) = u(L,t) = 0


```python
def solve_heat_equation(alpha, L, T, u0, nx, nt):
    """
    Решение уравнения теплопроводности методом конечных разностей (явная схема).
    """
    dx = L / (nx - 1)
    dt = T / nt
    
    x = np.linspace(0, L, nx)
    t = np.linspace(0, T, nt)
    
    # Проверка устойчивости
    r = alpha * dt / dx**2
    if r > 0.5:
        print(f"Предупреждение: коэффициент r = {r:.2f} > 0.5, схема может быть неустойчивой")
    
    # Инициализация решения
    u = np.zeros((nt, nx))
    u[0, :] = u0(x)
    
    # Явная схема
    for n in range(0, nt-1):
        for i in range(1, nx-1):
            u[n+1, i] = u[n, i] + r * (u[n, i+1] - 2*u[n, i] + u[n, i-1])
    
    return x, t, u
```

```python
# Пример использования
alpha = 0.01
L = 1
T = 1

def u0(x):
    return np.sin(np.pi * x / L)

nx = 50
nt = 1000

x, t, u = solve_heat_equation(alpha, L, T, u0, nx, nt)

# Визуализация
plt.figure(figsize=(10, 6))
for ti in range(0, nt, nt//10):
    plt.plot(x, u[ti, :], label=f't={t[ti]:.2f}')
plt.xlabel('x')
plt.ylabel('u(x,t)')
plt.title('Решение уравнения теплопроводности')
plt.legend()
plt.show()
```

## 4. Использование scipy.linalg.solve_banded для трехдиагональных матриц

Для повышения эффективности при работе с трехдиагональными матрицами:

```python
from scipy.linalg import solve_banded
```

```python
def solve_poisson_banded(f, a, b, alpha, beta, n):
    """
    Решение уравнения u''(x) = f(x) с использованием ленточных матриц.
    """
    h = (b - a) / (n - 1)
    x = np.linspace(a, b, n)
    
    # Инициализация ленточной матрицы (3 диагонали)
    ab = np.zeros((3, n))
    
    # Нижняя диагональ (i-1)
    ab[0, 2:] = 1/h**2
    # Главная диагональ (i)
    ab[1, :] = -2/h**2
    # Верхняя диагональ (i+1)
    ab[2, :-2] = 1/h**2
    
    # Граничные условия
    ab[1, 0] = 1
    ab[1, -1] = 1
    
    # Правая часть
    F = f(x)
    F[0] = alpha
    F[-1] = beta
    
    # Решение системы
    u = solve_banded((1, 1), ab, F)
    
    return x, u
```
